```python
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('../GlobalTemperatures.csv');
```

```python
ds.head()
```

```python
to_drop = ['LandMaxTemperature', 'LandMaxTemperatureUncertainty', 'LandMinTemperature', 'LandMinTemperatureUncertainty', "LandAndOceanAverageTemperature", "LandAndOceanAverageTemperatureUncertainty"]
ds.drop(to_drop, inplace=True, axis=1)
```

```python
ds.head()
```

```python
missing_data = ds.isnull().sum()
missing_data[0:10]
```

```python
subset_data = ds.loc[:, 'LandAverageTemperature':'LandAverageTemperatureUncertainty']
subset_data
```

```python
result_data = subset_data.fillna(0)
result_data.fillna(method='bfill', axis=0).fillna(0)
result_data.head(50)
```
